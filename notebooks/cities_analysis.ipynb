```python
import matplotlib.pyplot as plt

top10 = df.sort_values("population", ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top10["city"], top10["population"])
plt.xticks(rotation=45)
plt.title("Top 10 Ukrainian Cities by Population")
plt.show()
```

```python
import matplotlib.pyplot as plt

top10 = df.sort_values("population", ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.bar(top10["city"], top10["population"])

plt.title("Top 10 Ukrainian Cities by Population")
plt.xlabel("City")
plt.ylabel("Population")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
top10 = df.sort_values("population", ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.bar(top10["city"], top10["population"] / 1_000_000)

plt.title("Top 10 Ukrainian Cities by Population")
plt.xlabel("City")
plt.ylabel("Population (millions)")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
